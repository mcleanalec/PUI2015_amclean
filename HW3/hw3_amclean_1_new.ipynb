```python
%pylab inline
from __future__ import print_function
__author__= 'Alec'
import os
from scipy.optimize import curve_fit, minimize
import pylab as pl
import numpy as np

mymean=100
df = 100
md={}
distributions = ['chisq', 'norm', 'pois', 'binom', 'gamma', 'uni']

md['chisq'] = np.random.chisquare(df, size = 100)
md['pois'] = np.random.poisson(df, 100)
md['binom'] = np.random.binomial(200, .5, 100)
md['norm'] = np.random.normal(100, 1, 100)
md['gamma'] = np.random.gamma(100,1,100)
md['uni'] = np.random.uniform(100, 100,100)




mysize=2000/(np.array(range(1,100)+[10]))


```

```python
pl.hist(md['chisq'],bins=30)
pl.ylabel('N')
pl.xlabel('x')


md['chisq']={} 


for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size=n)
    
means = {}
means['chisq'] = {}
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_n = fig_mu_N.add_subplot(111)
for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    #and plot it
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
    
allmeans1=[]
for n in means['chisq']:
    allmeans1.append(means['chisq'][n])
pl.figure(figsize=(10,10))
pl.hist(allmeans1,bins=25, color='green')
pl.xlabel('sample mean')
pl.ylabel('N')
```
